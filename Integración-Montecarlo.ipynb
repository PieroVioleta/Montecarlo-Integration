# Integración usando numeros aleatorios

Este es el codigo usado para los resultados del trabajo de investigación sobre Integracion usando números aleatorios. Que fue anexado en la monografia del proyecto.

## Codigo Fuente

```python
%matplotlib inline

import matplotlib.pyplot as plt                    #importamos las librerias necesarias
import numpy as np
import random


def f(x):                                          #definimos la funcion a integrar
    return e**x



a = 0                                  #limite inferior de la integral
b = 2                                  #limite superior a integrar
alto_rect = np.e**2                     #alto del rectangulo que contiene la funcion

def area1(N): 
    plt.figure(figsize=(10,8))                  #tamaño de la figura
    x = np.random.uniform(a,b , N)              #generamos N valores aleatorios para x
    y = np.random.uniform(0, alto_rect, N)      #generamos N valores aleatorios para y
    debajo = y <= f(x)
    encima = np.invert(debajo)
    area_curva = (debajo.sum()*(b-a)*alto_rect)/N    #area debajo de la curva
    plt.plot(x[debajo], y[debajo], 'b.')
    plt.plot(x[encima], y[encima], 'r.')
    plt.axis([a,b,0,alto_rect])                     #definimos el rectangulo que contendra la funcion a integrar
    plt.plot(0,0,label= 'Area = {:4.5f}'.format(area_curva),alpha=0)
    plt.legend(frameon=True, framealpha=0.9, fontsize=25,loc='upper left')
```

## 1) Integral definida con antiderivada elemental

 Sea $$f(x) = e^x$$ Hallaremos  $$\int_0^2 \! f(x) \, \mathrm{d}x.$$ Veamos:

Para integrar las funciones basta con definir la función y cambiar los parametros **a, b y alto_rect**. Y si queremos aumentar la precisión basta con aumentar N. Y utilizar la funcion **area1(N).**

```python
def f(x):
    return np.e**x

a=0
b=2
alto_rect = np.e**2
N=5000

area1(N)
```

Aumentamos N a 100000, y obtenemos:

```python
N = 100000
area1(N)
```

## Codigo fuente del Histograma

```python
def hist(repeticiones):
    areas = []
    
    for i in range(repeticiones):
        xrand = np.random.uniform(a,b,N)
        sums = 0.0
    
        for i in range(N):
            sums += f(xrand[i])
    
        integral = ((b-a)/float(N))*sums
    
        areas.append(integral)

    plt.figure(figsize=(8,8))
    plt.title("Distribucion de las areas calculadas")
    plt.hist(areas, bins=30, ec='black')
    plt.xlabel("Areas")
    plt.ylabel("Numero de veces que se obtuvo el area")
```

Ahora repetiremos este experimento **repeticiones** veces extrayendo solo el area obtenida. Y ubicamos los datos en un histograma.
Basta con variar la variable **repeticiones** en la funcion **hist()** para ver los resultados del experimento y observar que a mas repeticiones se tiene un comportamiento normal. 

Realizando el experimento 100 veces tenemos:

```python
hist(repeticiones=100)
```

Ahora, si realizamos el experimento 2000 veces obtenemos:

```python
hist(repeticiones=2000)
```

Observamos que hay una tendencia a seguir una distribución Normal que es explicada en la monografia.
